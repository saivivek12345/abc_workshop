```python
import numpy as np
import matplotlib.pyplot as plt

# Samling
# step 1: Define the continuous time signal parameters
t_continuous = np.linspace(0, 1, 1000)  # Create a time array from 0 to 1 second with 1000 points for continuous time
frequency = 2  # Frequency of the sine wave in Hertz (5 cycles per second)
continuous_signal = np.sin(2 * np.pi * frequency * t_continuous)  # Generate a continuous sine wave signal

#plot continous signal
plt.plot(t_continuous, continuous_signal)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Continuous Time Signal')
plt.grid(True)
plt.show()
```

```python
# Read a Discrete signal values
O_signal = np.array([1, 3, 2, -1, -2, 0, 4, 3])
n = np.arange(len(O_signal))
# Plot Original Signal
plt.stem(n, O_signal, 'red')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Original Signal')
plt.grid(True)
plt.show()
```

```python
# Right shift (delaying the signal by 2 units)
shift_amount = 2
n_right_shifted = n + shift_amount
#plot the right time shifted signal
plt.stem(n_right_shifted, O_signal, 'red')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Right Time-Shifted Signal')
plt.grid(True)
plt.show()
```

```python
# Left shift (advancing the signal by 2 units)
shift_amount = 2
n_left_shifted = n - shift_amount
# Plot the Left Shift Signal
plt.stem(n_left_shifted, O_signal, 'red')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Left Time-Shifted Signal')
plt.grid(True)
plt.show()
```
